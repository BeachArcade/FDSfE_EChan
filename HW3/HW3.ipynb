**5.Load the USA state data**

```python
state_df = gpd.read_file("http://www2.census.gov/geo/tiger/GENZ2016/shp/cb_2016_us_state_5m.zip")
state_df.head()
```

**6.Plot Wisconsin State boundary**

```python
import matplotlib.pyplot as plt
fig, gax = plt.subplots(figsize=(10, 10))
state_df.query("NAME == 'Wisconsin'").plot(ax=gax, edgecolor="black", color="white")
plt.show()
```

**6.Load county data for Wisconsin state**

```python
county_df = gpd.read_file("http://www2.census.gov/geo/tiger/GENZ2016/shp/cb_2016_us_county_5m.zip")
county_df.head()
```

```python
# Only get Wisconsin Counties
county_df = county_df.query("STATEFP == '55'")
```

```python
#county_df.plot()

fig, gax = plt.subplots(figsize=(10, 10))


state_df.query("NAME == 'Wisconsin'").plot(ax=gax, edgecolor="black", color="white")



county_df.plot(ax=gax, edgecolor="black", color="white")

plt.show()
```

**7.Load voting results for 2016 between Trump and Clinton in Wisconsin state**

```python
results = pd.read_csv("https://datascience.quantecon.org/assets/data/ruhl_cleaned_results.csv", thousands=",")
results.head()
```

```python
results["county"] = results["county"].str.title()
results["county"] = results["county"].str.strip()
county_df["NAME"] = county_df["NAME"].str.title()
county_df["NAME"] = county_df["NAME"].str.strip()
```

**8.Spatially join the voting results and the county geometry**

```python

res_states = county_df.merge(results, left_on="NAME", right_on="county", how="inner")
```

```python
res_states.head()
```

```python
%%time
res_states["trump_share"] = res_states["trump"] / (res_states["total"])
res_states["rel_trump_share"] = res_states["trump"] / (res_states["trump"]+res_states["clinton"])
res_states.head()
```

```python
fig, gax = plt.subplots(figsize = (10,8))

# Plot the state

state_df[state_df['NAME'] == 'Wisconsin'].plot(ax = gax, edgecolor='black',color='white',figsize=(20,10))
# Plot the counties and pass 'rel_trump_share' as the data to color
res_states.plot(
    ax=gax, edgecolor='black', column='rel_trump_share', legend=True, legend_kwds={'label': "Republican Votes"}, cmap='RdBu_r',
    vmin=0.01, vmax=0.95
)

# Add text to let people know what we are plotting
gax.annotate('Republican vote share',xy=(0.76, 0.06),  xycoords='figure fraction')
gax.set_title("Republican Vote Share in WI")

# No axis with long and lat
plt.axis('off')
plt.show()
```

**9. How many counties did Trump win?**

```python
# Print the number of counties Trump won
print("Donald Trump won", res_states.eval("trump > clinton").sum(), "counties in the state of Wisconsin")
```

**9. How many counties did Clinton win?**

```python
# Print the number of counties Clinton won
print("Hilary Clinton won", res_states.eval("trump < clinton").sum(), "counties in the state of Wisconsin")
```

**10. How many votes did Trump had in total?**

```python
# Total of the Trump votes (1405284)
gop_votes = res_states.eval("trump").sum()
print('Trump had a total of', gop_votes, 'votes')
```

**11. How many votes did Clinton have in total?**

```python
# Total of the Clinton votes (1382536)
dem_votes = res_states.eval("clinton").sum()
print('Clinton had a total of', dem_votes, 'votes')
```

**12. Who won? and what was the vote difference?**
